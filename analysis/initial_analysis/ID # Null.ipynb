```python
#Null values per region
display(df.columns)
country_list = df['Country'].value_counts().index
null = pd.DataFrame()
countrylist = []
for country in country_list:
    df2 = df[df['Country']==country]
    to_append = df2.isna().sum().sort_values(ascending=False)
    null = null.append(to_append, ignore_index=True)
    countrylist.append(country)
```
